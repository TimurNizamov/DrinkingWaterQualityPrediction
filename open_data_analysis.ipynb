# Water compilance problems ration

```python
# ration negative compilance -> 1 water was ok to positive
years = df['year'].value_counts()
res = dict()
for year, value in dict(years).items():
    current_year = df[df['year']== year]
    compilance = current_year['compilance'].value_counts()
    res[year] = compilance[1]/sum(compilance)
res
```
