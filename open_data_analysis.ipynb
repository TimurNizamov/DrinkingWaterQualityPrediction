```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
def normalize_val(total, val):
    return val/total * 100

for index, year in enumerate(years):
    fig, ax = plt.subplots()
    ax.set_title(f'{year} Compilance ratio')
    ax.set_xticklabels([])
    
    cur_year_data = test.iloc[index]
    good = cur_year_data.compilance_1
    bad = cur_year_data.compilance_0
    
    total = good + bad
    
    good = normalize_val(total, good)
    bad = normalize_val(total, bad)
    
    a = plt.bar(-0.2, [good, 1], width=0.4, label = 'Good', color='green')
    b = plt.bar(+0.2, [bad, 1], width=0.4, label = 'Bad', color='red')
    ax.bar_label(a, label_type='center')

    ax.legend()
    plt.show()
```

# 3. Find interesting data by using apriori failed
Apriori failed here because of big data and a lot of diffirent features. Maybe apriori can find interesting facts if we continue digging data with it but after first 10 runs. It was clear it is preaty useless here.

```python
import apyori
def rules_to_df(rules):
    results = list(rules)
    apriori_df = pd.DataFrame(columns=('Items','Antecedent','Consequent','Support','Confidence','Lift'))

    Support =[]
    Confidence = []
    Lift = []
    Items = []
    Antecedent = []
    Consequent=[]
    print(Items)
    for RelationRecord in results:
        for ordered_stat in RelationRecord.ordered_statistics:
            Support.append(RelationRecord.support)
            Items.append(RelationRecord.items)
            Antecedent.append(ordered_stat.items_base)
            Consequent.append(ordered_stat.items_add)
            Confidence.append(ordered_stat.confidence)
            Lift.append(ordered_stat.lift)

    apriori_df['Items'] = list(map(set, Items))                                   
    apriori_df['Antecedent'] = list(map(set, Antecedent))
    apriori_df['Consequent'] = list(map(set, Consequent))
    apriori_df['Support'] = Support
    apriori_df['Confidence'] = Confidence
    apriori_df['Lift']= Lift
    return apriori_df
```

```python
copy = df.copy()
copy['water_tap_id'] = df['water_tap_id'].astype("string")
copy['year'] = df['year'].astype("string")
copy['compilance'] = df['compilance'].astype("string")
copy = copy.drop(['time_of_the_test', 'type_of_water'], axis=1)
copy = copy.fillna('-')
copy

interesting = copy[['water_tap_id', 'compilance_problem']]
col_names = interesting.values
rules = apyori.apriori(col_names, min_support = 0.01, min_confidence = 0.3)
# list(rules)
rules_df = rules_to_df(rules)
rules_df
```
