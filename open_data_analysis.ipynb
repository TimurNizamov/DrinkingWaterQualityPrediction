# Water tests and compilance by year

```python
df['year'].value_counts(normalize=True).plot(kind='bar', title='Performed water tests by year')
```

```python
test = pd.get_dummies(df,columns=['compilance'])
test = test.groupby(by='year', as_index=False).sum()
years = test.year
# test.loc[0]
test
```

```python
import matplotlib.pyplot as plt 
import numpy as np

def normalize_val(total, val):
    return val/total * 100

for index, year in enumerate(years):
    fig, ax = plt.subplots()
    ax.set_title(f'{year} Compilance ratio')
    ax.set_xticklabels([])
    
    cur_year_data = test.iloc[index]
    good = cur_year_data.compilance_1
    bad = cur_year_data.compilance_0
    
    total = good + bad
    
    good = normalize_val(total, good)
    bad = normalize_val(total, bad)
    
    
    
    a = plt.bar(-0.2, [good, 1], width=0.4, label = 'Good')
    b = plt.bar(+0.2, [bad, 1], width=0.4, label = 'Bad')
    ax.bar_label(a, label_type='center')
    
#     ax.text(3 + 3, 3 + .25, str(3), color='blue', fontweight='bold')

    ax.legend()
    plt.show()
```
