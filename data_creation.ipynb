<h1> 2. Extract required data from xml </h1>

<h2>2.1 Create Parser</h2>

```python
labels = ['veevark_id', 'veevark', 'proovivotu_eesmark', 'proovivotja_amet', 'proovivotja_nimi',
          'katseprotokollid', 'veeliik',  'proovivotu_aeg']

def extract_req_info(obj, labels):
    cur_obj = dict()
    for label in labels:
        cur_obj[label] = obj[label]
    return cur_obj

def extract_water_results(obj):
    try:
        quality_res = obj['katseprotokollid']['katseprotokoll']['hinnang']
    except TypeError:
        quality_res = obj['katseprotokollid']['katseprotokoll'][0]['hinnang']
    
    if quality_res == 'vastab nõuetele':
        return {'compilance': 1, 'compilance_problem': float('nan')}
    
    else:
        try:
            problem = obj['katseprotokollid']['katseprotokoll']['naitajad']['naitaja']
        except TypeError:
            problem = obj['katseprotokollid']['katseprotokoll'][0]['naitajad']['naitaja']

        return {'compilance': 0, 'compilance_problem': extract_water_problem(problem)}

def extract_water_problem(res_arr):
    problem = ''
    try:
        for obj in res_arr:
            if obj['hinnang'] != 'vastab nõuetele':
                problem += f"{str(obj['nimetus'])},"
        return problem[:-1]
    except TypeError:
        return res_arr['nimetus']


def extract_place(obj):
    return {'place': obj['proovivotukoht']['nimetus']}
    

def get_nice_dict(obj, year):
    cur_dict = extract_req_info(obj, labels)
    cur_dict.update(extract_water_results(cur_dict))
    cur_dict.update(extract_place(obj))
    cur_dict.pop('katseprotokollid')
    cur_dict['year'] = year
    return cur_dict
```

<h1>3. Create script to iterate through all files and create data frame </h1>

<h2> 3.1 Create dataframe from objects files </h2>

```python
years = ['2007', '2008', '2009', '2010', '2011', '2012',
         '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
clean_arr = []
for year in years:
    with open(f'xmls/{year}.xml', mode='rb') as f:
        data = xmltodict.parse(f)['veevargi_veeproovid']['proovivott']
    clean_arr = clean_arr + [get_nice_dict(obj, year) for obj in data]

df = pd.DataFrame(clean_arr)
df.head(10)
    
```
